```python
import spacy
import json
import os
import pandas as pd
import numpy as np
```

```python
nlp = spacy.load('ru_core_news_lg')
answers_path = "./utils/answers.json"
with open(answers_path) as f:
    answer_types = json.load(f)
```

```python
answer_types
```

```python
answers=[]
for key in answer_types:
    for item in answer_types[key]:
        answers.append(item)
```

```python
answers
```

```python
data = np.zeros((len(answers),len(answers)))
```

```python
df = pd.DataFrame(data = data, index=answers, columns = answers)
```

```python
df
```

```python
for i in answers:
    for j in answers:
        nlp_i = nlp(i)
        nlp_j = nlp(j)
        df.at[i,j] = nlp_i.similarity(nlp_j)
df
```

```python
df.to_csv('./result.csv')
```
